```python
class Student:
    # 생성자
    def __init__(self, name):
        self.name = name

    # 메서드
    def study(self, hard=False):
        if hard:
            print("%s 학생은 열심히 공부합니다." % self.name)
        else:
            print("%s 학생은 공부합니다." % self.name)

            s = Student('Heekyung')
            s = study()
            s.study(hard=True)
```

```python
import numpy as np # numpy 라이브러리 import

def print_val(x):
  # 'Type' 출력
  print("Type:")
  print(type(x))

  # 'Shape' 출력
  print("Shape:")
  # ndarray가 아닐 경우 shape이 없을 수 있으므로 try-except로 처리
  try:
    print(x.shape)
  except AttributeError:
    print("N/A (Not a numpy array)")

  # '값' 출력
  print("Value:")
  print(x)

# ---
```

```python
## In [2]
print_val(np.array([1,2,3])) # 1차원 배열

# ---

## In [5]
a = 5
print_val(a) # 일반 정수 (numpy 배열 아님)

# ---

## In [9] - np.array() 오류 수정됨
# 배열을 만들 때 리스트([ ]) 형태로 인자를 전달해야 합니다.
y = np.array([1, 2, 3, 4, 5, 6])
print_val(y)

# ---

## In [10]
y = np.ones((1,2))
print_val(y)

# ---

## In [11]
y = np.zeros((2,2))
print_val(y)

# ---

## In [12]
y = np.eye(3) # 3x3 항등 행렬
print_val(y)

# ---

## In [13] - np.eye() 오류 수정됨
# np.eye()는 보통 하나의 정수(차원)를 인자로 받습니다.
y = np.eye(2)
print_val(y)

# ---

## In [14] - np.ones() 오류 수정됨
# np.ones()도 튜플 형태로 shape을 지정하거나 단일 정수를 받습니다.
y = np.ones((2,2))
print_val(y)
```
